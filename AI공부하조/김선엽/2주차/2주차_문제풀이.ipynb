# 문제 1
**정답**

2번, 3번, 5번

**풀이**

(다중 선형 회귀는 MLR, 단순 선형회귀는 LR로 표기하겠다)

1. MLR은 보통 독립변수 갯수가 2개 이상을 가지고, LR은 독리변수가 1개이다.
따라서 1번 보기는 맞다
2. 선형회귀에서 가중치는 계수를 의미한다. 따라서 회귀에 사용하는 feature가 많아질 수록 
가중치의 개수는 증가하고, 사용하는 feature의 갯수가 하나라면 가중치의 개수는 1개이다.
LR의 가중치의 개수는 1개이기 때문에 2번 보기는 틀리다
3. 다항 회귀 분석에서는 독립변수와 종속변수의 관계가 비선형이지만, 선형 회귀 분석에서는 
독립변수와 종속변수의 관계는 선형이다. MLR, LR 모두 선형 회귀 분석이기 때문에 모두 독립변수와 종속변수의 관계는 선형이므로 3번 보기는 틀리다. 
4. MLR의 경우 feature의 개수가 여러개이기 때문에 B의 값은 여러개이다. 그리고 LR의 경우 feature의 개수가 한개이기 때문에 B의 값은 한개이다. 따라서 4번 보기는 맞다.
5. 행렬의 곱은 B의 값과 C의 값이 같아야 진행이 가능하다. 따라서 MLR의 B == C이어야 정상적인 
행렬 연산이 가능하므로 5번 보기는 틀리다

# 문제 2
**정답**

3번

**풀이**

1. 회귀 모델을 평가할 때 대표적으로 사용하는 R2 결정계수는 값이 1에 가까울수록 좋다고 본다. R2 결정계수의 값은 실제값과 예측값의 차이가 작을수록 값이 1에 가깝게 나온다. 따라서 모델을 실제값과 예측값의 차이가 작아지는 방향으로 학습을 시키는 것이 바람직하다는 것을 알 수 있다. 때문에 1번 선택지는 맞다.
2. MSE는 오차의 제곱의 평균을 사용하는 평가지표로 오차가 커질 수록 MSE의 값은 급격하게 증가하고, 오차가 작아질 수록  MSE의 값은 급격하게 감소한다. 따라서 MSE는 오차에 대해 민감하기 때문에 2번 선택지는 맞다.
3. R2 결정계수는 예측을 잘하면 값이 1에 가까워지고, 예측을 잘하지 못하면 값이 0에 가까워진다. 
여기서 오차는 (실제값 - 예측값)을 의미하는데, 오차가 줄어들 수록 R2 결정계수의 값은 0에 가까워진다. 즉 오차와 R2 결정계수는 반비례관계이므로 3번 선택지는 틀리다.
4. MAE는 실제값과 예측값의 절댓값 오차의 평균을 계산하는 방법이다.

# 문제 3

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

share_target = np.array([130, 128, 132, 140, 153, 150])
share_data = np.array([1, 3, 5, 7, 9, 11]).reshape(-1, 1)

y = np.dot(share_data, 2) + 125
reg = LinearRegression().fit(share_data, y)

for i in range(len(share_target)):
  print('%d 월 주가 = %d, 예측 주가 = %d'%(share_data[i] , share_target[i] ,reg.predict([share_data[i]])))

mse = mean_squared_error(share_target, y)
print('최종 mse:', mse)
print('12월 예상 주가 : %d'%(reg.predict([[12]])))


fig, ax = plt.subplots()
ax.plot(share_data, reg.coef_*share_data+reg.intercept_)
ax.scatter(share_data, share_target, color = 'red')
ax.set_xlabel('Month')
ax.set_ylabel('Cost')
```

# 문제 4

**정답**

5번

**풀이**

1. L1 Regularization은 Lasso를 L2 Regularization은 Ridge를 의미한다. 따라서 1번 선택지는 틀렸다.
2. 릿지와 라쏘는 규제 기법이기 이전에 선형회귀 모델이다.  선형회귀 모델은 변수 사이의 상관관계가 높은 데이터 일수록 예측력이 높기 때문에 두 방법 모두 예측력이 높다. 따라서 2번 선택지는 틀렸다.
3. RSS(잔차제곱합)이란 잔차(실제값과 예측값의 차이)의 제곱값을 더한 값을 의미한다. 여기에서 데이터의 갯수로 나누면 MSE가 된다. 
Ridge는 L2 norm을 사용하여 규제를 하고, Lasso는 L1 norm을 사용하여 규제를 한다. 하지만 L2 norm과 L1 norm 모두 계수를 이용하여 규제를 하지 RSS는 사용하지 않는다. 따라서 3번 선택지는 틀렸다.
4. Lasso 모델은 규제를 통해 모델에 불필요한 몇몇 계수를 0으로 만들기도 한다. 하지만 bias를 크게 한지는 않는다. 따라서 4번 선택지는 틀렸다.

계수를 가중치(weight)이라고도 하며, 절편을 bias라고도 하는 것 같다. bias가 어쩔 때는 편향을 의미하고 어쩔때는 절편을 의마한다니 헷갈려서 너무 싫다.

1. Ridge와 Lasso 모두 규제를 통해 과대적합을 방지한다. 하지만 모두 규제 강도를 조절하여 강도를 세게 함으로서 조금 더 과소적합을 유도하는 경향이 있다. 하지만 Ridge와 다르게 Lasso는 과소적합 가능성이 낮은 경향이 있고, 이 때문에 Lasso를 일반적으로 선호한다. 따라서 5번 선택지는 맞다.

# 문제 5

와인데이터 불러오기

```python
import pandas as pd
wine=pd.read_csv("http://bit.ly/wine-date")

wine.head()
wine.info()
```

시각화해보기

```python
import seaborn as sns
import matplotlib.pyplot as plt

#alcohol 히스토그램으로 그려보기
sns.histplot(data=wine, x="alcohol")
plt.show()

#alcohol 값의 분포를 boxplot으로 그려보기
sns.boxplot(x=wine['alcohol'])
plt.show()

#sugar 히스토그램으로 그려보기
sns.histplot(data=wine, x="sugar")
plt.show()

#sugar 값의 분포를 boxplot으로 그려보기
sns.boxplot(x=wine['sugar'])
plt.show()

#pH 히스토그램으로 그려보기
sns.histplot(data=wine, x="pH")
plt.show()

#pH 값의 분포를 boxplot으로 그려보기
sns.boxplot(x=wine['pH'])
plt.show()

#target값 확인해보기
g = sns.PairGrid(wine, hue="class")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

훈련데이터, 테스트데이터 나누기

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(wine.iloc[:, :3], wine.iloc[:, 3], random_state=42)
```

전처리

```python
from sklearn.preprocessing import StandardScaler

ss= StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

모델링

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)


print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))

# 예측
new = ([17, 10, 3.5] - ss.mean_) / ss.scale_
print(kn.predict([new]))
```
